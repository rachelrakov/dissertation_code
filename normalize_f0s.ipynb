```python
means = []
stds = []
medians = []
maxes = []
mins = []
for i, row in df_stats.iterrows():
    length = (len(row[1]))
    total = sum(row[1])
    mean = total/length
    std = np.std(row[1])
    mdn = stats.median(row[1])
    mx = max(row[1])
    mn = min(row[1])
    means.append(mean)
    stds.append(std)
    medians.append(mdn)
    maxes.append(mx)
    mins.append(mn)
    
print(len(means))
```
