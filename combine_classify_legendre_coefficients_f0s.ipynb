```python
all_f0s = []
for i, row in all_train_data.iterrows():
    f0s = np.fromstring(row["f0final"][2:-2], dtype=float, sep="\n")
    #print (type((f0s)))
    all_f0s.append(f0s)
```

```python
maxf0s = []
for listf0s in all_f0s:
    high = max(listf0s)
    maxf0s.append(high)
```

```python
minf0s = []
interlist = []
for listf0s in all_f0s:
    for item in listf0s:
        if item == 0:
            pass
        else:
            interlist.append(item)
    low = min(interlist)
    minf0s.append(low)
    interlist = []
```

```python
rangef0s = []
for i in range(len(maxf0s)):
    diff = maxf0s[i] - minf0s[i]
    rangef0s.append(diff)
print (len(rangef0s))
```

```python
df_f0s = pd.DataFrame({'maxf0s': maxf0s, 'minf0s':minf0s, 'rangef0s':rangef0s})
```
