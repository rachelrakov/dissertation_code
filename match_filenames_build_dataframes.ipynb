```python
import os
import pandas as pd
import glob
```

```python
#train files
train_ques_file = "C:/Users/Rachel/Documents/Grad_Stuff/Dissertation/train_data/format_train50txt/"
train_csvs = "C:/Users/Rachel/Documents/Grad_Stuff/Dissertation/train_data/trainf0csv/"
labels = "C:/Users/Rachel/Documents/Grad_Stuff/Dissertation/train_data/labels50_onlyqs.csv"
```

```python
#test files
train_ques_file = "C:/Users/Rachel/Documents/Grad_Stuff/Dissertation/test_data/format_testtxt/"
train_csvs = "C:/Users/Rachel/Documents/Grad_Stuff/Dissertation/test_data/testf0csv/"
labels = "C:/Users/Rachel/Documents/Grad_Stuff/Dissertation/test_data/test_labels_csv.csv"
```

```python
filenames = os.listdir(train_ques_file)
```

```python
languages = pd.read_csv(labels)
languages.iloc[0].values[0]
#languages.iloc[0].values[1]
```

```python
print (len(languages))
```

```python
df_all = []
for i, tqf in enumerate(sorted(filenames)):
    name, _= os.path.splitext(tqf)
    pattern = f"{train_csvs}{name}*"
    tqf_csv = sorted(glob.glob(pattern))
    try:
        questions = pd.read_csv(f'{train_ques_file}{tqf}', header=None)[4]
        f0final = [pd.read_csv(tq, delimiter=';')['F0final_sma'].values for tq in tqf_csv]
        lang = languages.iloc[i].values[0]
        #lang = languages.iloc[i].values[1]
    except pd.errors.EmptyDataError as e:
        print(name, e)
        continue
    except IndexError as x:
        print (name, x)
    
    try:
        df = pd.DataFrame({"file":name, "questions":questions, "f0final":f0final, "language":lang}, 
                          columns=["file", "language", "questions", "f0final"])
        df_all.append(df)
    except ValueError as e:
        print(e, "\n", name, f'questions:{len(questions)}', f'f0final:{len(f0final)}' )

        
    
```

```python
dfa = pd.concat(df_all, ignore_index=True)
```
