### How to get normal distributions of the legendre polynomials?
Group by language & question type
for each of the 4 categories

    - collect each individual polynomials - every first polynomial, every second poly nomial, etc. (into 8 different lists, one all [1,1,1,1,1 etc], one all [2,2,2,2,2 etc] and so on)

````python
coeffs = []
for i, row in all_train_data.iterrows():
    leges = np.fromstring(row["legen_coefs"][1:-1], dtype=float, sep=",")
    coeffs.append(leges)
````

````python
df_coeff = pd.DataFrame(coeffs, columns = [f'lg{n}' for n in range(7)])
````

````python
import matplotlib.pyplot as plt
x = np.linspace(-1,1, 100)
fig, ax = plt.subplots()
for (_, mean_row), (_, std_row) in zip(df_means_idx.iterrows(), df_std_idx.iterrows()):
    assert mean_row['language'] == std_row['language']
    assert mean_row['questions'] == std_row['questions']
    label = f"{mean_row['language']}: {mean_row['questions']}"
    fmean = np.poly1d(mean_row[2:].values)
    fstd = np.poly1d(std_row[2:].values)
    mean_line = fmean(x)
    std_line = fstd(x)
    _ = ax.plot(x, mean_line, label=label)
    _ = ax.fill_between(x, mean_line+std_line, mean_line-std_line, alpha=0.25)
_ = ax.legend(loc = "upper left")
````

````python
curve_probs = []
for i, row in df_coeff.iterrows():
    results = {'id':i, 'language': row['language'], 'questions':row['questions']}
    question_coeffs = row[:-2].values
    for (_, mean_row), (_, std_row) in zip(df_means_idx.iterrows(), df_std_idx.iterrows()):
        assert mean_row['language'] == std_row['language']
        assert mean_row['questions'] == std_row['questions']
        label = f"{mean_row['language']}: {mean_row['questions']}"
        mean_coeffs = mean_row[2:].values
        std_coeffs = std_row[2:].values
        log_sum = 0
        for qc, mc, stdc in zip(question_coeffs, mean_coeffs, std_coeffs):
            log_prob = norm.logpdf(qc, mc, stdc)
            log_sum +=log_prob
        final_prob = np.exp(log_sum)
        results[label] = final_prob
    curve_probs.append(results)
````

````python
df_final_probs = pd.DataFrame(curve_probs)
````
