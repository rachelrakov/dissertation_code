```python
degree = 7
ipu_coefs = []
for i, row in all_train_data.iterrows():
    start = row["start"]
    f0s = np.fromstring(row["f0final"][1:-1], dtype=float, sep=" ")
    x_time = []
    for j in range(len(f0s)):
        x_time.append(start+(0.01*j))
    #print (x_time[0], f0s[0])
    fit = L.fit(x_time, f0s, degree)
    #print(fit)
    coefs = [fit.coef[j] for j in range(degree)]
    #df = pd.DataFrame({"Legen_Coefs":ipu_coefs}, columns=["Legen_Coefs"])
    ipu_coefs.append(coefs)

    #df_leg.append(df)

```
