```python
f0_list = []
for i, row in data.iterrows():
    f0s = np.fromstring(row["f0final"][1:-1], dtype=float, sep=" ")
    #print (i, f0s)
    f0_list.append(f0s)
```

```python
#f0_list[0]
#print (len(ID_list))
print (len(f0_list))
```

```python
data["f0s"]=f0_list
```

```python
data = data.drop(labels="f0final", axis=1)
```

```python
data
```

```python
#### get the log of the f0s, put them into a list #####
interlist = []
log_f0s = []
for i, row in data.iterrows():
    f0s = row["f0s"]
    #print(f0s)
    for f0 in f0s:
        if f0 == 0.0:
            log_f0 = 0.00001
        else:
            log_f0 = math.log10(f0)
        interlist.append(log_f0)
    log_f0s.append(interlist)
    interlist = []
        
print(len(log_f0s))
```

```python
data["log_f0"]=log_f0s
data = data.drop(labels="f0s", axis=1)
```

```python
data
```

```python
df_stats = data
```

```python
for i, row in df_stats.iterrows():
    print(row[5])
    break
```

```python
means = []
stds = []
medians = []
maxes = []
mins = []
for i, row in df_stats.iterrows():
    length = (len(row[5]))
    total = sum(row[5])
    mean = total/length
    std = np.std(row[5])
    mdn = stats.median(row[5])
    mx = max(row[5])
    mn = min(row[5])
    means.append(mean)
    stds.append(std)
    medians.append(mdn)
    maxes.append(mx)
    mins.append(mn)
    
print(len(means))
```

```python
df_stats["means"]=means
df_stats['std']=stds
df_stats["medians"]=medians
df_stats["max"]=maxes
df_stats["min"]=mins
```

```python
df_stats
```

```python
interlist = []
f0_final = []
for i,row in df_stats.iterrows():
    mean = row["means"]
    log_f0s = row["log_f0"]
    for i in log_f0s:
        norm_f0 = i - mean
        interlist.append(norm_f0)
    f0_final.append(interlist)
    interlist = []
print(len(f0_final))
```

```python
data["f0final"] = f0_final
data = data.drop(labels="log_f0", axis=1)
```
