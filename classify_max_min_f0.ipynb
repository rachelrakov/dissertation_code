```python
all_f0s = []
for i, row in all_train_data.iterrows():
    f0s = np.fromstring(row["f0final"][2:-2], dtype=float, sep="\n")
    #print (type((f0s)))
    all_f0s.append(f0s)

    #leges = np.fromstring(row["legen_coefs"][1:-1], dtype=float, sep=",")
    #coeffs.append(leges)
```

```python
maxf0s = []
#minf0s = []
for listf0s in all_f0s:
    high = max(listf0s)
    #low = min(listf0s)
    maxf0s.append(high)
    #minf0s.append(low)

 
```

```python
minf0s = []
interlist = []
for listf0s in all_f0s:
    for item in listf0s:
        if item == 0:
            pass
        else:
            interlist.append(item)
    low = min(interlist)
    minf0s.append(low)
    interlist = []
```
